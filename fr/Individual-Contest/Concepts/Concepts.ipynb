# Concepts

## **1. Description du probleme**

**Concepts** est un jeu de devinette de mots dans lequel les joueurs communiquent des idees au moyen d'icones visuelles. Il existe deux roles : le **donneur d'indices** et le **devineur**. Les deux joueurs disposent d'un ensemble commun d'icones visuelles, chacune accompagnee d'une description connue. Voici quelques icones d'exemple :

<img src="../../../Individual-Contest/Concepts/figs/Concepts Fig 1.png" width="300">

Le donneur d'indices choisit d'abord un **secret**, c'est-a-dire un mot ou une expression, puis fournit un **indice** en pointant une **sequence ordonnee** d'icones issues de l'ensemble partage : il n'est pas permis de parler ni d'ecrire.

L'ordre des icones dans l'indice est significatif :

- La **premiere icone** represente generalement l'idee centrale du secret.
- Les **icones suivantes** apportent un contexte supplementaire qui aide a clarifier ou a developper le concept principal.

### Exemple 1

L'indice suivant peut etre interprete comme *un lieu ou se deroule un metier qui combat le feu*, autrement dit une **caserne de pompiers** :

<img src="../../../Individual-Contest/Concepts/figs/Concepts Fig 2.png" width="300">

Si l'ordre des icones est inverse, il peut plutot suggerer *un metier qui combat le feu dans une maison*, ce qui renvoie a un **pompier** :

<img src="../../../Individual-Contest/Concepts/figs/Concepts Fig 3.png" width="300">

### Exemple 2

Les icones peuvent prendre des sens differents selon leur contexte. Par exemple, l'icone coeur peut apparaitre dans l'indice suivant, qui suggere *un outil utilise par les medecins pour ecouter le coeur*, c'est-a-dire un **stethoscope** :

<img src="../../../Individual-Contest/Concepts/figs/Concepts Fig 4.png" width="300">

La meme icone coeur peut au contraire apparaitre dans un indice qui implique *un personnage fictif a la fois mort et vivant*, ce qui renvoie a un **zombie** :

Lors de l'etape a domicile, les concurrents avaient developpe un programme d'IA capable de predire les reponses a partir d'une sequence d'indices. C'etait amusant. Mais votre ami vous lance maintenant un nouveau defi :  
*"Deviner, c'est facile ; mais pouvez-vous aussi creer un systeme d'IA capable de donner un bon indice ?"*

En fait, il vous met encore davantage au defi : pouvez-vous creer un systeme d'IA qui fournisse un bon indice afin qu'un autre systeme d'IA puisse deviner votre mot-cle ?

Pour rendre le defi plus interessant, nous jouons maintenant a la version habituelle du jeu Concepts. Pour rappel, notre version precedente etait simplifiee de sorte qu'un indice ne pouvait contenir qu'une seule sequence de marqueurs.

Vous pouvez maintenant fournir jusqu'a **4 sequences de marqueurs** !

Cela permet d'exprimer plus clairement des idees complexes. Voici un exemple utilisant 3 sequences de marqueurs pour expliquer un **samourai** :

**Nouveau format alimente par l'IA :**
Nous avons remplace le devineur humain par un **devineur IA**. Pour fluidifier le jeu :

1. Le mot cible (*label*) est toujours choisi dans un ensemble predefini (`options`).
2. Les indices doivent etre une **sequence ordonnee de marqueurs** choisis exclusivement parmi un ensemble fixe de **118 marqueurs candidats**.

**Terminologie**

Afin d'assurer clarte et coherence, nous normalisons les termes essentiels dans tous les supports.

- **label** : la reponse cible (ou le "secret") a identifier.
- **options** : l'ensemble predefini de candidats parmi lesquels tous les *labels* valides sont choisis.
- **marker** : une icone representant un concept, accompagnee de sa description textuelle.
- **hints** : une **sequence ordonnee** de *markers* fournie au devineur IA pour l'aider a identifier le *label*.

Votre tache consiste a fournir des **hints** pour chaque *label* afin d'aider le **devineur IA** a l'identifier. Par exemple, lorsque le label cible est `"microphone"`, votre programme peut generer quatre sequences d'indices comme suit :

1.  **Indice 1** :  
    `["Object-Box", "Electronic-Computing", "Mouth-Taste", "Ear-Sound-Hearing", "Tool-Construction"]`
2.  **Indice 2** :  
    `["Music-Song", "Television-Program-Show", "Work-Occupation", "Use - Action-Do - Verbe-Button"]`
3.  **Indice 3** :  
    `["Black", "Metal", "Plastic-Rubber", "Cylinder", "Circle-Ring"]`
4.  **Indice 4** :  
    `["Arm-Hand-Finger", "Happy-Positive", "Expression - Quote-Talking-Words", "Life-Heart-Love"]`

**Remarque** : l'exemple ci-dessus est donne uniquement a titre illustratif. Le programme doit generer des listes d'ID et non des chaines de caracteres ; reportez-vous a la fois a la section `3. Tache` et au [baseline.ipynb](https://ioai.bohrium.com/notebooks/26681337682).

## **3. Tache**

Votre tache consiste a fournir un **indice** pour chaque *label* afin d'aider le **devineur IA** a l'identifier. Vous devez developper un programme qui prend deux entrees :

1. Une chaine `label` (votre *label* secret)
2. La liste `options` (100 choix candidats pour le devineur)

Le `label` est garanti comme appartenant a `options`.

De plus, le programme utilise les `candidate markers` predefinis.

<img src="../../../Individual-Contest/Concepts/figs/Concepts Fig 8.png" width="600">

Ce programme doit retourner, pour chaque `label`, une liste de listes d'entiers representant les indices (c'est-a-dire les sequences de marqueurs). Plus precisement :

- La liste retournee doit contenir **au plus 4** sequences.
- Chaque sequence peut contenir **jusqu'a 8 entiers**.
- Chaque entier represente l'ID d'un marqueur dans l'indice.

Vos indices sont ensuite fournis a un devineur IA en boite noire. Vous marquez un point si ce devineur identifie correctement votre mot-cle secret a partir de vos indices.

## **5. Evaluation**

Votre indice est evalue avec deux metriques :

### Hits@10

= 1 si le mot secret figure parmi les 10 premieres propositions de l'IA, sinon 0.

### NDCG@10 (Normalized Discounted Cumulative Gain)  
Recompense davantage une bonne reponse lorsqu'elle apparait plus haut dans la liste.

Si le mot secret se trouve au rang *i* (en commencant a 1) :

$$
\text{NDCG@10} = \frac{1}{\log_2(i + 1)}
$$

**Exemples :**
- Rang 1 -> 1.00  
- Rang 2 -> ~0.63  
- Rang 4 -> ~0.43  
- Rang 10 -> ~0.29

### Score final

Votre score final est une combinaison des deux mesures : 0.9 Hits@10 + 0.1 NDCG@10.

Cela signifie que vous obtenez deja un score important des que le devineur trouve correctement le mot-cle secret, mais qu'un score plus eleve est attribue si le devineur le predit plus tot.

## **7. Exigences**

### Exigences generales

- Limite maximale de soumissions : **15 fois**. Seules les soumissions reussies (c'est-a-dire celles qui recoivent un score sur le classement A) sont comptabilisees.
- Contraintes de l'environnement de test : la machine de test execute votre notebook en **10 minutes** maximum. Si l'execution depasse **10 minutes**, le systeme l'interrompt de force et renvoie un retour de type "Timeout" ou "Failed".
- Soumission de donnees et de modeles : dans cette tache, les concurrents doivent soumettre un notebook et peuvent joindre 1 jeu de donnees cree par eux-memes. Sa taille totale ne doit pas depasser `2GB`. Joindre un tres gros jeu de donnees peut allonger le processus de test, car il doit etre monte sur la machine d'inference. Ce temps ne compte pas dans la limite d'execution, mais il rallonge le delai avant de pouvoir consulter et selectionner les soumissions.
- Reseau : lors de l'etape sur site, la machine de test ne peut pas acceder a Internet. Autrement dit, les commandes de telechargement comme 'pip' et 'conda', ainsi que les appels API, ne fonctionnent pas. Pour ce probleme en particulier, les machines de test ne peuvent acceder ni a l'API du devineur ni au proxy LLM.

### Limites d'acces a l'API

Votre jeton vous donne droit a 12,500 requetes `POST` vers l'endpoint `/guess` de l'API du devineur IA.  

- En cas de succes, l'API renvoie un dictionnaire de la forme `{'guesses': ['firefighter', 'fire inspector', 'fire marshal', 'fire warden', 'fire safety officer', 'smokejumper', 'pyrotechnician', 'firewatcher', 'chef', 'cook'], 'message': 'Generated 10 guesses'}`.  
- En cas d'echec, elle leve soit une `HTTPStatusError` et renvoie `{'detail': 'error message'}`, soit `{'guesses': [], 'message': 'unable to generate guesses'}`.  
- En general, le premier cas signifie que vous avez depasse la limite de 12,500 requetes de votre jeton, ou que vous avez effectue plus de 1000 requetes en 1 minute ; le second signifie que vos indices depassent 4 sequences ou 8 marqueurs dans une sequence.  
- Reportez-vous au message d'erreur pour plus de details.
- En periode de charge extremement elevee (attendue dans la derniere heure du concours), il peut y avoir 1 a 2 echecs pour 1000 requetes ; un mecanisme de nouvelle tentative est fourni dans [baseline.ipynb](https://ioai.bohrium.com/notebooks/26681337682).

**Remarque sur le determinisme des sorties**

Dans le bloc de code ci-dessus, nous fixons la graine aleatoire afin de garantir la reproductibilite des resultats lorsque le modele baseline est execute seul. Il s'agit d'une pratique courante pour eliminer la variance des sorties causee par les operations stochastiques.
Cependant, dans cette tache precise, votre modele doit interagir dynamiquement avec un devineur IA. Celui-ci repose sur un grand modele de langage, qui peut produire des reponses differentes pour une meme entree en raison de l'aleatoire inherent a son processus de decodage. Ainsi, meme avec le meme ensemble d'indices, les reponses du devineur peuvent varier d'une execution a l'autre.
Par consequent, vous pouvez observer des fluctuations dans les scores rapportes entre plusieurs soumissions identiques. C'est un comportement attendu et non un bogue du systeme d'evaluation.
